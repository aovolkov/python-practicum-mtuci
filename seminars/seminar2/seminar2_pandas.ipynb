#### Объединение данных

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7']})


frames = [df1, df2]

result = pd.concat(frames)
result
````

````python
result = pd.concat([df1, df3], axis=1, join='inner')
result
````

````python
result = df1.append(df2)
result
````

#### Работа с категориальными признаками: 
One hot encoder - используется для преобразования категориальных переменных в числовые значения. В этом процессе каждое целочисленное значение представляется как двоичный вектор, равный нулю, за исключением индекса целого числа, который помечен 1.
Источник: https://pythonpip.ru/osnovy/one-hot-encoding-python

````python
df = pd.DataFrame({'cat1': ['a', 'b', 'c', 'b', 'd', 'a'],
                   'cat2': ['a1', 'b1', 'c1', 'b1', 'd1', 'a1'],
                  'cat3': np.arange(6)})
df
````

````python
pd.get_dummies(df)
````

````python
pd.get_dummies(df['cat3'])
````

#### Работа с категориальными признаками:
Label encoder - каждое значение кодируется натуральным числом. 

````python
codes, uniques = pd.factorize([20,10,np.nan,10,np.nan,30,20])
print(codes)
print(uniques)
````

````python
codes, uniques = pd.factorize([20,10,np.nan,10,np.nan,30,20], 
                              na_sentinel=None)
print(codes)
print(uniques)
````

### Pandas vs SQL

#### WHERE

````python
url = ('https://raw.github.com/pandas-dev/pandas/'
       'master/pandas/tests/io/data/csv/tips.csv')
tips = pd.read_csv(url)
tips.head()
````

Запрос из SQL:

```
SELECT *
FROM tips
WHERE time = 'Dinner'
LIMIT 5;
```
С помощью pandas:

````python
tips[tips['time'] == 'Dinner'].head(5)
````

Запрос из SQL (чаевые более $5,00 за ужин):

```
SELECT *
FROM tips
WHERE time = 'Dinner' AND tip > 5.00;
```
С помощью pandas:

````python
tips[(tips['time'] == 'Dinner') & (tips['tip'] > 5.00)]
````

Запрос из SQL (чаевые от толиков состоящих не менее чем из 5 человек, ИЛИ общая сумма счета превышает 45 долларов):

```
SELECT *
FROM tips
WHERE size >= 5 OR total_bill > 45;
```
С помощью pandas: 

````python
tips[(tips['size'] >= 5) | (tips['total_bill'] > 45)]
````

#### JOIN

````python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': np.random.randn(4)})

df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],
                    'value': np.random.randn(4)})
````

````python
df1
````

````python
df2
````

Запрос из SQL:

```
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
```
С помощью pandas: 

````python
pd.merge(df1, df2, on='key')
````

Запрос из SQL (показать все записи из df1):

```
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
```
С помощью pandas: 

````python
pd.merge(df1, df2, on='key', how='left')
````

Запрос из SQL (показать все записи из df2):

```
SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;
```
С помощью pandas: 

````python
pd.merge(df1, df2, on='key', how='right')
````

Запрос из SQL (показать все записи из обеих таблиц):

```
SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;
```
С помощью pandas: 

````python
pd.merge(df1, df2, on='key', how='outer')
````

### Группировка

Запрос из SQL:

```
SELECT sex, count(*)
FROM tips
GROUP BY sex;
```
С помощью pandas: 

````python
tips.groupby('sex').size()
````

````python
tips.groupby('sex').count()
````

````python
tips.groupby('sex')['total_bill'].count()
````

Запрос из SQL:

```
SELECT day, AVG(tip), COUNT(*)
FROM tips
GROUP BY day;
```
С помощью pandas: 

````python
tips.groupby('day').agg({'tip': np.mean, 'day': np.size})
````

````python
tips.groupby('day').agg({'tip': [np.mean, np.min], 'day': np.size})
````

Запрос из SQL:

```
SELECT smoker, day, COUNT(*), AVG(tip)
FROM tips
GROUP BY smoker, day;
```
С помощью pandas: 

````python
tips.groupby(['smoker', 'day']).agg({'tip': [np.size, np.mean]})
````
